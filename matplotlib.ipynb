```python
a= np.linspace(0,10,11)
```

```python
a
```

```python
b=a**4
```

```python
b
```

```python
x = np.arange(0,10)
```

```python
y=x*2
```

```python
y
```

```python
fig = plt.figure()
#Large
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_title('Power of 4')

#Small
axes2 = fig.add_axes([0.2, 0.4, 0.5, 0.5])
axes2.plot(x,y)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_xlim(0,5)
axes2.set_ylim(0,10)
axes2.set_title('Zoom in')

plt.show()
```

```python
fig= plt.figure(figsize=(12,8), dpi=600)
axes1 =fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, dpi=200)

axes[0][0].plot(x,y)
axes[1][1].plot(a,b)

fig.subplots_adjust(wspace=0.9, hspace=0.9)

fig.suptitle('Figure Level', fontsize=16)
fig.tight_layout()
```

```python
fig = plt.figure()

ax= fig.add_axes([0,0,1,1])

ax.plot(x,x,color= 'green',lw=3,ls='-.', label ='X vs X')

lines=ax.plot(x,x**2,color= 'red',lw=5,marker= 'o',markersize=10, label= 'X vs X^2')
lines[0].set_dashes([5,2,5,2])

ax.legend(loc =(1.1, 0.5))
```
